# revgeo() - Used for mapping latitude and longitude to states, street,city,zip,country

```r
locations <- na.omit(locations)
rev <- revgeo(longitude = locations$lon, latitude = locations$lat, output = "frame", item="state")
```

```r
rev
```

```r
library(dplyr)
```

# Summarizing the state and count of tweets from that state using group_by() and summarize()

```r
statelist <- rev[rev$country == "United States of America", ]
levels(statelist$state) <- tolower(levels(statelist$state))
# tweetfreq<-as.data.frame(table(states))
statelist <- statelist %>% mutate(freq = 1)
statelist <- statelist %>% group_by(state) %>% summarize(sum(freq))
```

```r
statelist
```

```r
USAstates <- data.frame(state = tolower(rownames(USArrests)))
TotalUsaStates <- merge(x = USAstates , y = statelist , by = "state", all.x = TRUE)
```

```r
TotalUsaStates
```

```r
#Assigning 0 to states freqeuncy whose freq data is NA
TotalUsaStates[is.na(TotalUsaStates)] <-0
```

```r
TotalUsaStates
```

# A simple data package to ease the process of creating choropleths in 'ggplot2' with all fifty U.S. states and Washington D.C., including Alaska and Hawaii as insets.

```r
library(ggplot2)
library(fiftystater)

data("fifty_states") 

colnames(TotalUsaStates)[2] <-"Freq"
# map_id creates the aesthetic mapping to the state name column in your data
p <- ggplot(TotalUsaStates, aes(map_id = state)) + 
  # map points to the fifty_states shape data
  geom_map(aes(fill = Freq ), map = fifty_states, color="black") + 
  expand_limits(x = fifty_states$long, y = fifty_states$lat) +
  coord_map() +
  scale_x_continuous(breaks = NULL) + 
  scale_y_continuous(breaks = NULL) +
  labs(x = "", y = "") + 
    ggtitle("Flu Tweets Analysis by State") + 
  theme(legend.position = "bottom", 
        panel.background = element_blank())

p <- p+scale_fill_gradient(low="white", high="blue")
```

```r
p
```

```r
saveRDS(p, file = "p.rds")
```

```r
cdcData<-read.csv("StateDatabyWeekforMap_2018-19week40-5.csv")
cdcData$level<-as.integer(gsub("[^0-9\\.]", "", cdcData$ACTIVITY.LEVEL))
```

```r
fluData <- data.frame(state = tolower(cdcData$STATENAME), cdcData$level)
colnames(fluData)[2] <- "countLevel"

fluData <- fluData %>% group_by(state) %>% summarize(ceiling(mean(countLevel)))
colnames(fluData)[2] <- "countLevel"
fluData <- na.omit(data.frame(fluData))

# map_id creates the aesthetic mapping to the state name column in your data
p1 <- ggplot(fluData, aes(map_id = state)) + 
  # map points to the fifty_states shape data
  geom_map(aes(fill = countLevel), map = fifty_states, color="black") +  
  expand_limits(x = fifty_states$long, y = fifty_states$lat) +
  coord_map() +
  scale_x_continuous(breaks = NULL) + 
  scale_y_continuous(breaks = NULL) +
  labs(x = "", y = "") +
  ggtitle(" CDC Flu Data Analysis")+
  theme(legend.position = "bottom", 
        panel.background = element_blank(), plot.title = element_text(size=9))
p1 <-p1 +scale_fill_gradient(low="white", high="red")

p1
#install.packages("gridExtra")
#require(gridExtra)

#p2 <- grid.arrange(p, p1, ncol=2)

saveRDS(p1, file = "p1.rds")
```
